# Recommendation system

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Devopia/youtube_urls_by_topic1.csv')
```

```python
df['Topic'] = df['Topic'].replace('-', '', regex=True)
```

```python
df.head(15)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(ngram_range=(1,2))

tfidf = vectorizer.fit_transform(df["Topic"])
```

## main

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def search(Topic):
    query_vec = vectorizer.transform([Topic])
    similarity = cosine_similarity(query_vec, tfidf).flatten()
    indices = np.argpartition(similarity, -5)[-5:]
    results = df.iloc[indices].iloc[::-1]

    return results
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

vectorizer = TfidfVectorizer(ngram_range=(1,2))
tfidf = vectorizer.fit_transform(df["Topic"])

list_of_topic = ['mensuration', 'fractions']
for topic in list_of_topic:
  query_vec = vectorizer.transform([topic])
  similarity = cosine_similarity(query_vec, tfidf).flatten()
  indices = np.argpartition(similarity, -5)[-5:]
  results = df.iloc[indices].iloc[::-1]
  topic_url_dict = results.groupby('Topic')['URL'].apply(list).to_dict()
  print(topic_url_dict)
```

```python
import ipywidgets as widgets
from IPython.display import display

topic_input = widgets.Text(
    value='',
    description='Topic:',
    disabled=False
)
topic_list = widgets.Output()
list_of_topics = ['mensuration']

def on_type(data):
    with topic_list:
        topic_list.clear_output()
        for topic in list_of_topics:
          title = topic
          if len(title) > 1:
            urls = search(title)
            topic_url_dict = urls.groupby('Topic')['URL'].apply(list).to_dict()
            print(topic_url_dict)

topic_input.observe(on_type, names='value')


display(topic_input, topic_list)
```
